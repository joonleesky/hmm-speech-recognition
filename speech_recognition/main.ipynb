## HMM Properties

```python
hmm_dict = {}
cnt = 0
state_num = 3
hmm_txt = open('hmm.txt')

#model 개수만큼 반복
for num in range(21):

    #알파벳 할당
    word = hmm_txt.readline().split('"')[1]  #
    hmm_dict[word] = {}   
    #BEGIN 
    hmm_txt.readline()                  
                 
    num_state = hmm_txt.readline().split()     #State 개수
    hmm_dict[word][num_state[0]] = num_state[1] 
    
    if num == 20:
        state_num = 1
    #STATES
    for st in range(state_num):

        state = hmm_txt.readline().split()[1] #State 할당
        hmm_dict[word][state] = {}   
    
        num_mix = hmm_txt.readline().split() #mixture 개수
        hmm_dict[word][state][num_mix[0]] = num_mix[1]  


        #MIXTURES
        hmm_dict[word][state]['<MIXTURES>'] = {}      #Mixture 별 Properties

        for mix in range(10):
            #MIXTURE
            mixture = hmm_txt.readline().split()
            hmm_dict[word][state]['<MIXTURES>'][mixture[1]] = {}
            hmm_dict[word][state]['<MIXTURES>'][mixture[1]][mixture[0]] = mixture[2]

            #MEAN
            hmm_txt.readline()  #mean개수는 39개로 통일임
            mean = hmm_txt.readline().split()   #mean은 numpy 형태로 저장       
            hmm_dict[word][state]['<MIXTURES>'][mixture[1]]['<MEAN>'] = mean

            #Variance
            hmm_txt.readline()  
            var = hmm_txt.readline().split()   #mean은 numpy 형태로 저장     
            hmm_dict[word][state]['<MIXTURES>'][mixture[1]]['<VARIANCE>'] = var

            #GCONIST
            gcon = hmm_txt.readline().split()  
            hmm_dict[word][state]['<MIXTURES>'][mixture[1]][gcon[0]] = gcon[1]
    
    #Transpose
    hmm_txt.readline()
    trans = []
    
    #'sp'는 3개의 state
    if(num == 20):
        for k in range(3):
            trans.append(hmm_txt.readline().split())
    else:  #5개의 state
        for k in range(5):
            trans.append(hmm_txt.readline().split())
    hmm_dict[word]['TRANSP'] = trans
    #END
    hmm_txt.readline()
```

```python
def gmm_prob(x,mixtures):
    b = []
    sum_exp = 0
    for i in range(1,11):       #10개의 mixture
        gconst = float(mixtures['%d'%i]['<GCONST>'])
        mixture = float(mixtures['%d'%i]['<MIXTURE>'])
        mean = mixtures['%d'%i]['<MEAN>']
        variance = mixtures['%d'%i]['<VARIANCE>']
        
        mean = np.array(list(map(float,mean)), dtype=np.float64)
        variance = np.array(list(map(float,variance)), dtype=np.float64)
        
        #gconst2 = np.log(np.prod(np.sqrt(variance))*np.power((2*np.pi),19.5))*2   이거랑 같은 값 나옴

        b_i = (-1/2)*np.sum(np.power((x-mean)/np.sqrt(variance),2))-gconst/2+np.log(mixture)
        b.append(b_i)
    for i in range(10):
        exp = b[i]-np.max(b)
        sum_exp += np.exp(exp)
        
    return np.max(b) + np.log(sum_exp)
```

```python
def cur_state_probability(test,component,state_seq):
    a = np.log(float(hmm_dict[component]['TRANSP'][state_seq+1][state_seq+1]))  #다음 state로 밖에 이동하지 못한다
    b = gmm_prob(test,hmm_dict[component][str(state_seq+2)]['<MIXTURES>']) #다음 state에서의 b 값
    return a+b
    
def next_state_probability(test,component,state_seq):
    a = np.log(float(hmm_dict[component]['TRANSP'][state_seq+1][state_seq+2]))
    b = gmm_prob(test,hmm_dict[component][str(state_seq+3)]['<MIXTURES>']) #다음 state에서의 b 값
    return a+b

def skip_state_probability(test,component,state_seq): #state_seq =0 : state2->4로 가야됨
    a = np.log(float(hmm_dict[component]['TRANSP'][state_seq+1][state_seq+3]))  
    b = gmm_prob(test,hmm_dict[component][str(state_seq+4)]['<MIXTURES>']) #다음 state에서의 b 값
    return a+b

def back_state_probability(test,component,state_seq): #state_seq =2 : state4->2로 가야됨
    a = np.log(float(hmm_dict[component]['TRANSP'][state_seq+1][state_seq-1]))  
    b = gmm_prob(test,hmm_dict[component][str(state_seq)]['<MIXTURES>']) #다음 state에서의 b 값
    return a+b

def state_4_next_probability(test,component_cur,component_next,state_seq):
    a = np.log(float(hmm_dict[component_cur]['TRANSP'][3][4]))   #state '4' -> 나갈 확률
    b = gmm_prob(test,hmm_dict[component_next]['2']['<MIXTURES>']) #다음 component에서의 mixture 확률
    return a+b
```

```python
def sp_next(max_matrix,word,state,test,time):  #state한개만 존재
    prev = V[time-1][state]
    max_matrix[state][state] = prev + cur_state_probability(test,'sp',0)  #현재로 돌 확률
    component = 'sp'

    #다음 단어로 이동할 확률
    cnt = 0
    for next_word,components in word_dict.items():
        next_state = start_point[cnt]   #다음 word의 state번호
        cnt +=1
        a_0_j = float(bi_gram[word][next_word])

        if a_0_j == 0: 
            continue
        b = gmm_prob(test,hmm_dict[components[0]]['2']['<MIXTURES>'])
        max_matrix[state][next_state] = prev + np.log(a_0_j) +b
    
    return max_matrix
        
#sp가 아닐때의 확률 계산
def not_sp(max_matrix,word,state,start_point,test,time):  #ex)word = 'eight', state = 8, start_point = 3
    component_val = (state-start_point) // 3   # ex) state 13, start 10 -> 2번째의 state'2'이다.
    state_seq = (state-start_point) %  3
    
    prev = V[time-1][state]
    
    if (state_seq < 2):
        component = word_dict[word][component_val]
        max_matrix[state][state] = prev + cur_state_probability(test,component,state_seq)
        max_matrix[state][state+1] = prev + next_state_probability(test,component,state_seq)
    else:
        component_cur = word_dict[word][component_val]
        component_next = word_dict[word][component_val+1]      #다음 component로 이동했을 때의 확률도 계산
        max_matrix[state][state] = prev + cur_state_probability(test,component_cur,state_seq)
        max_matrix[state][state+1] = prev + state_4_next_probability(test,component_cur,component_next,state_seq)
        
    return max_matrix
```

```python
def sil_next(max_matrix,state,test,time):
    prev = V[time-1][state]
    
    state_seq = state %  3    #sil안에는 component 항상 1개
    max_matrix[state][state] = prev + cur_state_probability(test,'sil',state_seq)  #현재로 돌아올 확률

    if state_seq == 0:   #2번 state
        max_matrix[state][state+1] = prev + next_state_probability(test,'sil',state_seq)  #2번->3번
        max_matrix[state][state+2] = prev + skip_state_probability(test,'sil',state_seq)  #2번->4번
    elif state_seq == 1:  #3번 state
        max_matrix[state][state+1] = prev + next_state_probability(test,'sil',state_seq)  #3번->4번
    elif state_seq == 2:  #4번 state
        max_matrix[state][state-2] = prev + back_state_probability(test,'sil',state_seq)  #2번->4번
        
        cnt = 0
        component = 'sil'
        for next_word,components in word_dict.items():
            next_state = start_point[cnt]   #다음 word의 state번호
            cnt +=1
            a_0_j = float(bi_gram['<s>'][next_word])
            if a_0_j == 0: 
                continue
            b = gmm_prob(test,hmm_dict[components[0]]['2']['<MIXTURES>'])
            max_matrix[state][next_state] = prev + np.log(a_0_j) +b
    
    return  max_matrix
```

```python
def find_word(state):
    cnt = 0
    for start in start_point:
        if state - start < 0:
            return word_list[cnt-1]
        cnt +=1
    return 'zero'
```

```python
word_list = ['<s>','eight','five','four','nine','oh','one','seven','six','three','two','zero']

def step(V,time):    #ex) word:eight, components:'ey, V:V , time:1
    max_matrix = np.zeros([113,113])
    for state in range(0,113):
        if V[time-1][state] == 0:
            continue

        if(state in sp):    #sp
            word = find_word(state)
            max_matrix = sp_next(max_matrix, word ,state,test[time],time)
            
        else:
            if state in sil:   #sil
                max_matrix = sil_next(max_matrix, state, test[time],time)

                
            elif((state-3) >= 0) & ((state-10) < 0): #eight
                max_matrix = not_sp(max_matrix, 'eight', state, 3, test[time],time)

            elif((state-10) >= 0) & ((state-20) < 0):  #five
                max_matrix = not_sp(max_matrix, 'five', state, 10, test[time],time)
                
            elif((state-20) >= 0) & ((state-30) < 0):  #four
                max_matrix = not_sp(max_matrix, 'four', state, 20, test[time],time)
                
            elif((state-30) >= 0) & ((state-40) < 0):  # nine
                max_matrix = not_sp(max_matrix, 'nine', state, 30, test[time],time)
                
            elif((state-40) >= 0) & ((state-44) < 0):  #oh
                max_matrix = not_sp(max_matrix, 'oh', state, 40, test[time],time)
                
            elif((state-44) >= 0) & ((state-54) < 0):  #one
                max_matrix = not_sp(max_matrix, 'one', state, 44, test[time],time)
                
            elif((state-54) >= 0) & ((state-70) < 0):  #seven
                max_matrix = not_sp(max_matrix, 'seven', state, 54, test[time],time)
                
            elif((state-70) >= 0) & ((state-83) < 0):  #six
                max_matrix = not_sp(max_matrix, 'six', state, 70, test[time],time)
                
            elif((state-83) >= 0) & ((state-93) < 0):  #three
                max_matrix = not_sp(max_matrix, 'three', state, 83, test[time],time)
                
            elif((state-93) >= 0) & ((state-100) < 0):  #two
                max_matrix = not_sp(max_matrix, 'two', state, 93, test[time],time)
                
            else:
                max_matrix = not_sp(max_matrix, 'zero', state, 100, test[time],time)
    
    max_matrix = np.where(max_matrix == 0 ,-100000 ,max_matrix)
    max_arg = np.argmax(max_matrix,0)
    max_matrix = np.max(max_matrix,0)
    max_matrix = np.where(max_matrix == -100000 ,0 ,max_matrix)
    return max_matrix,max_arg
```
